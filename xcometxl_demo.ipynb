```python
# Your Windows 11 course translations
source_text_1 = "In this course, Windows 11 Essentials, we will focus on important aspects of Windows 11 features and functionality including the File Explorer, built-in Windows Apps (such as Calendar, Mail, Photos, Teams), tablet tools, keyboard shortcuts and more."

target_text_1 = "Dans ce cours, Windows 11 Essentials, nous nous concentrerons sur les aspects importants des fonctionnalités de Windows 11, notamment l'explorateur de fichiers, les applications Windows intégrées (telles que Calendrier, Mail, Photos, Teams), les outils de tablette, les raccourcis clavier et plus encore."

source_text_2 = "After creating a new folder, Sasha noticed that if she waits a minute the folder will pop up to the top of the documents folder. Why does it do this?"

target_text_2 = "Après avoir créé un nouveau dossier, Sasha a remarqué qu'en attendant une minute, le dossier s'affichait en haut du dossier Documents. Pourquoi ?"

source_text_3 = "Demonstrate organizational techniques for file management"

target_text_3 = "Démontrer des techniques d'organisation pour la gestion des fichiers"

# Prepare data for XCOMET
data = [
    {"src": source_text_1, "mt": target_text_1},
    {"src": source_text_2, "mt": target_text_2},
    {"src": source_text_3, "mt": target_text_3}
]

print(f"Prepared {len(data)} translation pairs for evaluation")
```

```python
print("Running XCOMET-XL evaluation...")

# Run prediction with L4-optimized settings
model_output = model.predict(data, batch_size=8, gpus=1)

print("Evaluation complete!\n")
print("=" * 60)
print("SEGMENT-LEVEL QUALITY SCORES")
print("=" * 60)

for i, score in enumerate(model_output.scores, 1):
    quality = "Excellent" if score >= 0.8 else "Good" if score >= 0.6 else "Moderate" if score >= 0.4 else "Poor"
    print(f"Segment {i}: {score:.4f} ({quality})")

print(f"\nSYSTEM-LEVEL SCORE: {model_output.system_score:.4f}")
```

```python
# @title
print("\n" + "=" * 60)
print("DETAILED ERROR ANALYSIS")
print("=" * 60)

error_spans = model_output.metadata.error_spans

for i, (segment_errors, segment_data) in enumerate(zip(error_spans, data), 1):
    print(f"\n{'-' * 60}")
    print(f"SEGMENT {i} (Score: {model_output.scores[i-1]:.4f})")
    print(f"{'-' * 60}")

    src_preview = segment_data['src'][:80] + '...' if len(segment_data['src']) > 80 else segment_data['src']
    mt_preview = segment_data['mt'][:80] + '...' if len(segment_data['mt']) > 80 else segment_data['mt']

    print(f"\nSource: {src_preview}")
    print(f"Target: {mt_preview}")

    if segment_errors and len(segment_errors) > 0:
        print(f"\n{len(segment_errors)} error(s) detected:")
        for j, error in enumerate(segment_errors, 1):
            print(f"\n  Error {j}:")
            print(f"    - Text: '{error.get('text', 'N/A')}'")
            print(f"    - Severity: {error.get('severity', 'N/A')}")
            print(f"    - Type: {error.get('type', 'N/A')}")
            if 'start' in error and 'end' in error:
                print(f"    - Position: chars {error['start']}-{error['end']}")
    else:
        print("\nNo errors detected - excellent translation!")
```

```python
from rich.table import Table
from rich.console import Console

console = Console()

# Create a summary table
table = Table(title="XCOMET-XL Translation Quality Report", show_header=True, header_style="bold magenta")
table.add_column("Segment", style="cyan", justify="center")
table.add_column("Score", justify="center")
table.add_column("Quality", justify="center")
table.add_column("Errors", justify="center")
table.add_column("Severity", justify="left")

for i, (score, errors) in enumerate(zip(model_output.scores, error_spans), 1):
    # Determine quality rating
    if score >= 0.8:
        quality = "[green]Excellent[/green]"
    elif score >= 0.6:
        quality = "[yellow]Good[/yellow]"
    elif score >= 0.4:
        quality = "[orange]Moderate[/orange]"
    else:
        quality = "[red]Poor[/red]"

    # Count errors by severity
    error_count = len(errors) if errors else 0
    severities = [e.get('severity', 'unknown') for e in errors] if errors else []
    severity_str = ", ".join(set(severities)) if severities else "None"

    table.add_row(
        f"{i}",
        f"{score:.4f}",
        quality,
        str(error_count),
        severity_str
    )

console.print(table)
print(f"\nOverall System Score: {model_output.system_score:.4f}")
```

```python
def evaluate_translations(source_list, target_list, batch_size=8):
    """
    Evaluate multiple translation pairs.

    Args:
        source_list: List of source texts
        target_list: List of target texts
        batch_size: Batch size (default 8 for L4)

    Returns:
        Dict with scores and statistics
    """
    data = [{"src": src, "mt": tgt} for src, tgt in zip(source_list, target_list)]

    print(f"Evaluating {len(data)} translations...")
    model_output = model.predict(data, batch_size=batch_size, gpus=1)

    scores = model_output.scores
    results = {
        "scores": scores,
        "system_score": model_output.system_score,
        "error_spans": model_output.metadata.error_spans,
        "average_score": sum(scores) / len(scores),
        "min_score": min(scores),
        "max_score": max(scores),
        "excellent_count": sum(1 for s in scores if s >= 0.8),
        "good_count": sum(1 for s in scores if 0.6 <= s < 0.8),
        "moderate_count": sum(1 for s in scores if 0.4 <= s < 0.6),
        "poor_count": sum(1 for s in scores if s < 0.4),
    }

    print(f"Evaluation complete!")
    print(f"Average Score: {results['average_score']:.4f}")
    print(f"  Excellent (>=0.8): {results['excellent_count']}")
    print(f"  Good (0.6-0.8): {results['good_count']}")
    print(f"  Moderate (0.4-0.6): {results['moderate_count']}")
    print(f"  Poor (<0.4): {results['poor_count']}")

    return results
```

```python
from pydantic import BaseModel, Field
from typing import List

class TranslationSegment(BaseModel):
    """A single aligned segment of source and target text."""
    segment_id: int = Field(description="Sequential ID starting from 1")
    source_text: str = Field(description="Source text segment, exactly as provided, no modifications")
    target_text: str = Field(description="Target translation segment, exactly as provided, no modifications")
    word_count: int = Field(description="Approximate word count of this segment")
    reasoning: str = Field(description="Brief explanation of why this segmentation boundary was chosen")

class SegmentationResponse(BaseModel):
    """Complete segmentation of source and target texts."""
    segments: List[TranslationSegment] = Field(description="List of aligned translation segments")
    total_segments: int = Field(description="Total number of segments created")
    segmentation_strategy: str = Field(description="Overall strategy used for segmentation")

print("Pydantic models defined successfully!")
```

```python
import os
import json
from openai import OpenAI
from google.colab import userdata

def intelligent_segment_for_xcomet(
    source_text: str,
    target_text: str,
    max_words_per_segment: int = 400,
    openrouter_api_key: str = None
) -> SegmentationResponse:
    """
    Intelligently segment source and target texts using LLM reasoning.

    Args:
        source_text: Full source text (English)
        target_text: Full target text (French/other language)
        max_words_per_segment: Maximum words per segment (default 400 for XCOMET)
        openrouter_api_key: OpenRouter API key (or set OPENROUTER_API_KEY env var)

    Returns:
        SegmentationResponse with aligned segments
    """
    # Get API key
    api_key = userdata.get("OPENROUTER_API_KEY")
    if not api_key:
        raise ValueError("OpenRouter API key required. Set OPENROUTER_API_KEY env var or pass as argument.")

    # Initialize OpenAI client with OpenRouter
    client = OpenAI(
        base_url="https://openrouter.ai/api/v1",
        api_key=api_key
    )

    # Comprehensive system prompt
    system_prompt = f"""You are an expert text segmentation specialist for translation quality evaluation.

YOUR TASK:
Segment the provided source text and its translation into aligned chunks suitable for quality evaluation. Each segment must be a maximum of {max_words_per_segment} words.

CRITICAL RULES - YOU MUST FOLLOW THESE EXACTLY:

1. NEVER MODIFY THE TEXT
   - Return text EXACTLY as provided
   - Do NOT correct translations
   - Do NOT fix typos or grammar
   - Do NOT rephrase or improve anything
   - Your job is ONLY to segment, not to translate or correct

2. MAINTAIN PERFECT ALIGNMENT
   - Each source segment MUST have a corresponding target segment
   - Segments must cover the same semantic content
   - Number of source segments MUST equal number of target segments

3. SEGMENT INTELLIGENTLY
   - Respect natural section breaks (headings, paragraphs, lessons)
   - Keep semantically related content together
   - If a section exceeds {max_words_per_segment} words, split it at:
     * Sentence boundaries (preferred)
     * Paragraph breaks
     * Logical thought transitions
   - Avoid splitting mid-sentence unless absolutely necessary

4. WORD COUNT LIMITS
   - Maximum {max_words_per_segment} words per segment
   - Count words in the longer of source/target for each segment
   - If a single sentence exceeds limit, split at clause boundaries

5. HANDLE DIFFERENT CONTENT TYPES
   - Course descriptions: Keep as single segment if under limit
   - Lesson titles + descriptions: Keep together if under limit
   - Quiz questions: Keep question + all answers together if possible
   - Lists: Keep related items together

6. PROVIDE CLEAR REASONING
   - Explain why you chose each segmentation boundary
   - Note any difficult segmentation decisions
   - Justify splits within long sections

EXAMPLE SCENARIOS:

Scenario 1: Clear section breaks
Source: "Title 1\\n\\nContent 1. Content 2.\\n\\nTitle 2\\n\\nContent 3."
Action: Split at section breaks (before each title)

Scenario 2: Long section
Source: A 600-word paragraph
Action: Split at sentence boundary near 400-word mark, explain in reasoning

Scenario 3: Short sections
Source: Three 150-word sections
Action: Keep as separate segments or combine if semantically related

Remember: Your segmentation directly impacts translation quality evaluation. Poor segmentation = inaccurate quality scores."""

    # User prompt with both texts
    user_prompt = f"""Segment these texts for translation quality evaluation:

SOURCE TEXT (English):
{source_text}

TARGET TEXT (Translation):
{target_text}

Create segments with maximum {max_words_per_segment} words each. Maintain perfect source-target alignment.
Return EXACT text - do not modify, correct, or improve anything."""

    # Convert Pydantic model to JSON schema
    schema = {
        "type": "object",
        "properties": {
            "segments": {
                "type": "array",
                "items": {
                    "type": "object",
                    "properties": {
                        "segment_id": {"type": "integer"},
                        "source_text": {"type": "string"},
                        "target_text": {"type": "string"},
                        "word_count": {"type": "integer"},
                        "reasoning": {"type": "string"}
                    },
                    "required": ["segment_id", "source_text", "target_text", "word_count", "reasoning"]
                }
            },
            "total_segments": {"type": "integer"},
            "segmentation_strategy": {"type": "string"}
        },
        "required": ["segments", "total_segments", "segmentation_strategy"]
    }

    print("Calling OpenRouter with Gemini 2.5 Flash (with reasoning enabled)...")

    # Make API call with structured output and reasoning
    response = client.chat.completions.create(
        model="google/gemini-2.5-flash",
        temperature=0.3,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        response_format={
            "type": "json_schema",
            "json_schema": {
                "name": "segmentation_response",
                "schema": schema,
                "strict": False
            }
        },
        extra_body={
           "reasoning": {
            "enabled": True
        }
        }  # Enable reasoning for better segmentation decisions
    )

    # Parse response
    result_json = json.loads(response.choices[0].message.content)

    # Convert to Pydantic model
    segmentation_result = SegmentationResponse(**result_json)

    print(f"Segmentation complete! Created {segmentation_result.total_segments} segments.")

    # Show reasoning if available
    if hasattr(response.choices[0].message, 'reasoning'):
        print(f"\nModel Reasoning: {response.choices[0].message.reasoning}")

    return segmentation_result

print("Intelligent segmentation function ready!")
```

```python
# @title
from pydantic import BaseModel, Field
from typing import List, Optional

class ErrorContext(BaseModel):
    """Context about a specific error from XCOMET"""
    error_text: str = Field(description="The problematic text span")
    error_type: str = Field(description="Type of error (e.g., Mistranslation, Omission)")
    severity: str = Field(description="Severity level (minor, major, critical)")

class CorrectionRequest(BaseModel):
    """Input data for correction"""
    segment_id: int
    source_text: str
    target_text: str
    quality_score: float
    errors: List[ErrorContext]
    word_count: int

class CorrectionResponse(BaseModel):
    """Corrected translation output"""
    segment_id: int = Field(description="ID of the segment being corrected")
    corrected_translation: str = Field(description="The corrected translation, faithful to source")
    correction_reasoning: str = Field(description="Explanation of what was corrected and why")
    changes_made: List[str] = Field(description="List of specific changes made")
    confidence: float = Field(description="Confidence in correction quality (0.0-1.0)")

print("✓ Correction models defined!")
```

```python
import os
import json
from openai import OpenAI
from google.colab import userdata

def correct_translation_segment(
    correction_request,
    openrouter_api_key: str = None
) -> CorrectionResponse:
    """
    Intelligent translation correction agent using Gemini 2.5 Pro with reasoning.

    Args:
        correction_request: CorrectionRequest with source, target, and XCOMET diagnostics
        openrouter_api_key: OpenRouter API key

    Returns:
        CorrectionResponse with corrected translation
    """
    api_key = userdata.get("OPENROUTER_API_KEY")
    if not api_key:
        raise ValueError("OpenRouter API key required")

    client = OpenAI(
        base_url="https://openrouter.ai/api/v1",
        api_key=api_key
    )

    # Build error context summary
    error_details = "\n".join([
        f"  • Error: '{err.error_text}' | Type: {err.error_type} | Severity: {err.severity}"
        for err in correction_request.errors
    ])

    # Comprehensive system prompt for Tier 3 correction
    system_prompt = """You are an expert translation correction specialist for educational content.

YOUR ROLE: Tier 3 Correction Agent
You receive translations that have been flagged by XCOMET-XL (a neural quality estimation system) as having accuracy or fluency issues. Your job is to provide corrected translations that are:

1. FAITHFUL TO SOURCE: Must accurately convey the meaning of the original English text
2. NATURAL & FLUENT: Must sound natural in the target language
3. CONTEXTUALLY APPROPRIATE: Must fit the educational content domain
4. TERMINOLOGICALLY CONSISTENT: Must use correct technical terms

CRITICAL RULES:

1. PRESERVE MEANING
   - Your correction MUST match the source text meaning exactly
   - Do NOT add information not in the source
   - Do NOT omit information from the source
   - Do NOT change the intent or tone

2. FIX IDENTIFIED ERRORS
   - You are provided with specific errors detected by XCOMET
   - Address each flagged error in your correction
   - Explain what you fixed and why

3. MAINTAIN NATURALNESS
   - Correct translations should sound natural to native speakers
   - Use appropriate register for educational content
   - Maintain consistent terminology

4. BE SURGICAL
   - If only one phrase has an error, correct that phrase
   - Don't unnecessarily change correct portions
   - Preserve good translations while fixing errors

5. PROVIDE CLEAR REASONING
   - Explain each correction you make
   - Reference the specific XCOMET errors you addressed
   - Note if you made additional improvements beyond flagged errors

EDUCATIONAL CONTENT CONTEXT:
- This is technical training material (Windows 11 course)
- Clarity and accuracy are paramount
- Students rely on these translations for learning
- Technical terms must be precise

XCOMET ERROR TYPES YOU MAY ENCOUNTER:
- Mistranslation: Wrong word/phrase used
- Omission: Source content missing in translation
- Addition: Extra content not in source
- Terminology: Incorrect technical term
- Grammar: Grammatical errors
- Fluency: Awkward or unnatural phrasing

Remember: A critical mistranslation (like "triangular" instead of "tetrahedral" in chemistry) can completely mislead learners. Your corrections directly impact educational quality."""

    user_prompt = f"""SEGMENT TO CORRECT:

Segment ID: {correction_request.segment_id}
Quality Score: {correction_request.quality_score:.4f}
Word Count: {correction_request.word_count}

SOURCE TEXT (English):
{correction_request.source_text}

CURRENT TRANSLATION (needs correction):
{correction_request.target_text}

XCOMET DIAGNOSTIC REPORT:
{error_details}

TASK:
Provide a corrected translation that:
1. Accurately reflects the source text meaning
2. Addresses all errors identified by XCOMET
3. Maintains natural fluency in the target language
4. Uses appropriate technical terminology

Return your correction with clear reasoning about what you fixed. confidence is a score between 0 and 1"""

    # JSON schema for structured output
    schema = {
        "type": "object",
        "properties": {
            "segment_id": {"type": "integer"},
            "corrected_translation": {"type": "string"},
            "correction_reasoning": {"type": "string"},
            "changes_made": {
                "type": "array",
                "items": {"type": "string"}
            },
            "confidence": {"type": "number"}
        },
        "required": ["segment_id", "corrected_translation", "correction_reasoning", "changes_made", "confidence"]
    }

    # Make API call with reasoning enabled
    response = client.chat.completions.create(
        model="x-ai/grok-code-fast-1",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        response_format={
            "type": "json_schema",
            "json_schema": {
                "name": "correction_response",
                "schema": schema,
                "strict": True
            }
        },
    )

    # Parse response
    try:
        result_json = json.loads(response.choices[0].message.content)
    except json.JSONDecodeError as e:
        print(f"JSON Decode Error: {e}")
        print("Raw LLM response content:")
        print(response.choices[0].message.content)
        raise ValueError(f"LLM did not return valid JSON. Check console for raw output. Error: {e}") from e

    correction_result = CorrectionResponse(**result_json)

    return correction_result
```

```python
# @title
console.print(Panel.fit(
    "[bold magenta]Re-evaluation: Measuring Improvement[/bold magenta]\n"
    "[dim]Running XCOMET on corrected translations[/dim]",
    border_style="magenta",
    padding=(1, 2)
))
console.print()

if corrections:
    # Prepare corrected segments for XCOMET
    corrected_segments_for_xcomet = [
        {
            "src": corr['request'].source_text,
            "mt": corr['response'].corrected_translation
        }
        for corr in corrections
    ]

    console.print("[yellow]⚙ Running XCOMET evaluation on corrected segments...[/yellow]\n")

    # Re-evaluate with XCOMET
    corrected_output = model.predict(corrected_segments_for_xcomet, batch_size=1, gpus=1)

    console.print("[green]✓ Re-evaluation complete![/green]\n")

    # Compare results
    comparison_table = Table(
        title="[bold]Before vs After Correction[/bold]",
        box=box.ROUNDED,
        show_header=True,
        header_style="bold magenta",
        show_lines=True
    )
    comparison_table.add_column("Seg", style="dim", width=4, justify="center")
    comparison_table.add_column("Before", justify="center", width=10)
    comparison_table.add_column("After", justify="center", width=10)
    comparison_table.add_column("Δ Change", justify="center", width=10)
    comparison_table.add_column("Errors Before", justify="center", width=8)
    comparison_table.add_column("Errors After", justify="center", width=8)
    comparison_table.add_column("Status", width=15)

    improvements = []

    for i, (corr, new_score) in enumerate(zip(corrections, corrected_output.scores)):
        old_score = corr['original_score']
        delta = new_score - old_score

        old_errors = len(corr['request'].errors)
        new_errors = len(corrected_output.metadata.error_spans[i]) if corrected_output.metadata.error_spans[i] else 0

        # Determine status
        if delta > 0.05:
            status = "[bold green]✓ Improved[/bold green]"
            status_color = "green"
        elif delta < -0.05:
            status = "[bold red]✗ Worse[/bold red]"
            status_color = "red"
        else:
            status = "[yellow]≈ Similar[/yellow]"
            status_color = "yellow"

        # Color code scores
        before_color = "red" if old_score < 0.7 else "yellow" if old_score < 0.9 else "green"
        after_color = "red" if new_score < 0.7 else "yellow" if new_score < 0.9 else "green"
        delta_color = "green" if delta > 0 else "red" if delta < 0 else "yellow"

        comparison_table.add_row(
            str(corr['request'].segment_id),
            f"[{before_color}]{old_score:.4f}[/{before_color}]",
            f"[{after_color}]{new_score:.4f}[/{after_color}]",
            f"[{delta_color}]{delta:+.4f}[/{delta_color}]",
            f"[red]{old_errors}[/red]",
            f"[green]{new_errors}[/green]" if new_errors < old_errors else f"[yellow]{new_errors}[/yellow]",
            status
        )

        improvements.append({
            'segment_id': corr['request'].segment_id,
            'before_score': old_score,
            'after_score': new_score,
            'delta': delta,
            'before_errors': old_errors,
            'after_errors': new_errors
        })

        # Update the correction dict with new score
        corr['new_score'] = new_score
        corr['improvement'] = delta

    console.print(comparison_table)
    console.print()

    # Summary stats
    avg_improvement = sum(imp['delta'] for imp in improvements) / len(improvements)
    improved_count = sum(1 for imp in improvements if imp['delta'] > 0.05)
    worse_count = sum(1 for imp in improvements if imp['delta'] < -0.05)

    summary_table = Table(box=box.ROUNDED, show_header=True, header_style="bold cyan")
    summary_table.add_column("Metric", style="cyan")
    summary_table.add_column("Value", justify="right", style="white")

    summary_table.add_row("Segments Corrected", str(len(corrections)))
    summary_table.add_row("Improved", f"[green]{improved_count}[/green]")
    summary_table.add_row("Worse", f"[red]{worse_count}[/red]")
    summary_table.add_row("Similar", f"[yellow]{len(corrections) - improved_count - worse_count}[/yellow]")
    summary_table.add_row("Average Improvement", f"[{'green' if avg_improvement > 0 else 'red'}]{avg_improvement:+.4f}[/]")

    console.print(Panel(
        summary_table,
        title="[bold magenta]Correction Impact Summary[/bold magenta]",
        border_style="magenta"
    ))
    console.print()

else:
    console.print("[dim]No segments were corrected[/dim]\n")

print("✓ Re-evaluation complete")
```
